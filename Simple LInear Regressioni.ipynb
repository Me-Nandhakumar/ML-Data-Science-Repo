# Trend Analysis of Gold Prices in India (1990–2026) Using Linear Regression Models

```python
import pandas as pd
```

```python
data = {
    "Year": [
        1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,
        2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,
        2019,2020,2021,2022,2023,2024,2025,2026],
    "Gold_Price_10g_INR": [
        4234,4400,4300,4990,5600,5850,7000,8400,10800,12500,
        14500,18500,26400,31050,29600,28006.5,26343.5,
        28623.5,29667.5,31438,35220,48651,48720,52670,
        65330,77913,130000,143400]
}
df = pd.DataFrame(data)
print(df)
```

# Feature Selection/ Engineering

```python
x=df[['Year']]
x
y=df.iloc[:,1]
y
```

# Model Train/Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train) 
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color='GREEN', label='Actual Marks')
plt.plot(x_train, lr.predict(x_train), color='ORANGE', linewidth=2, label='Regression Line')
plt.title('Year vs Gold_Price_10g_INR')
plt.xlabel('Year')
plt.ylabel('Gold_Price_10g_INR')
plt.legend()
plt.show()
```
